## 1.5 ETL — Extract, Transform, Load

```python
# EXTRACT: Check raw data quality
print('=== EXTRACT: RAW DATA QUALITY CHECK ===')
print(f'Total Records  : {len(df):,}')
print(f'Total Columns  : {len(df.columns)}')
print(f'Missing Values : {df.isnull().sum().sum()}')
print(f'Duplicate Rows : {df.duplicated().sum()}')
```

```python
# TRANSFORM: Clean and prepare data
print('=== TRANSFORM: CLEANING DATA ===')

# Remove duplicates
df_clean = df.drop_duplicates()
print(f'Rows after removing duplicates: {len(df_clean):,}')

# Drop ID column (not needed for analysis)
df_clean = df_clean.drop(columns=['ID'])
print(f'Columns after dropping ID: {df_clean.columns.tolist()}')

# Rename target column for clarity
df_clean = df_clean.rename(columns={'Reached.on.Time_Y.N': 'Late_Delivery'})
print('\n✅ Target column renamed to: Late_Delivery')
print('   (1 = Late, 0 = On Time)')
```

```python
# LOAD: Save cleaned data
print('=== LOAD: SAVING CLEANED DATA ===')

# Create outputs folder if not exists
os.makedirs('../outputs', exist_ok=True)

# Save cleaned dataset
df_clean.to_csv('../outputs/cleaned_data.csv', index=False)
print('✅ Cleaned data saved to: outputs/cleaned_data.csv')
print(f'Final Shape: {df_clean.shape[0]:,} rows, {df_clean.shape[1]} columns')
```
